Data visualization

```python
cancelled_pre = df['is_canceled'].value_counts(normalize=True)
```

```python
cancelled_pre
```

```python
plt.figure(figsize=(4,5))
plt.title('show the  parcentage of is_canceled column ')
plt.bar(['Not canceled', 'canceled'],df['is_canceled'].value_counts(normalize=True), edgecolor='r', width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.title("Reservation status in diffrent Hotal")
axs= sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = axs.get_legend_handles_labels()
axs.legend(['Not canceled', 'Canceled'],fontsize=6)
plt.xlabel('Hotal')
plt.ylabel('Number of reservations')

plt.show()
```

```python
resort_hotel = df[df['hotel']== 'Resort Hotel']
city_hotel = df[df['hotel']== 'City Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby(['reservation_status_date'])[['adr']].mean()
city_hotel = city_hotel.groupby(['reservation_status_date'])[['adr']].mean()
```

```python
plt.figure(figsize=(10,4))
plt.title('Avrage daily rate in city and resort')
plt.plot(resort_hotel.index,resort_hotel['adr'], label="Resort Hotel")
plt.plot(city_hotel.index,city_hotel['adr'], label="city Hotel")
plt.legend()
plt.xlabel('Date')
plt.ylabel('Avrage daily rate')
plt.grid()
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month

df['month']
```

```python
plt.figure(figsize= (8,4))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='Reds')
plt.legend([ 'Not canceled','Canceled'])
plt.title('MOnth wise booking cancellation')
plt.xlabel('Month')
plt.ylabel('Number of bookings')
plt.show()
```

```python
# plt.figure(figsize=(10,5))
sns.barplot(x='month', y='adr',data = df[df['is_canceled'] == 1].groupby(['month'])[['adr']].mean().reset_index(),color='red')
plt.title('Month ke according data')
plt.xlabel('Month')
plt.legend(['is_Canceled'])
plt.ylabel('Avrage daily rate')
plt.show()
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.pie(top_10_country, autopct= '%.2f%%', labels = top_10_country.index  )

plt.title(' Canceled data of top 10 Country')
plt.legend(top_10_country)
plt.show()
```

```python
canceled = df[df['is_canceled'] == 1]
print(canceled['reservation_status_date'])
canceled['market_segment'].value_counts()



```

```python
canceled_df_adr= canceled.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date', inplace=True)


not_canceled_df_adr= df[df['is_canceled'] == 0]
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr= not_canceled_df_adr.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize=(10,4))
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='Canceled')
plt.plot(not_canceled_df_adr.index, not_canceled_df_adr['adr'], label='Not Canceled')
print(not_canceled_df_adr.index)
plt.title('Avrage daily rate in canceled and not canceled')
plt.legend()
plt.show()
```
